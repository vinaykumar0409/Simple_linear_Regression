```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data
area = np.array([1400, 1600, 1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700])
price = np.array([245000, 312000, 279000, 308000, 199000, 219000, 405000, 324000, 319000, 255000])
```

```python
def simple_linear_regression(x, y) :
  mean_x = np.mean(x)
  mean_y = np.mean(y)
  m = np.sum((x - mean_x) * (y - mean_y)) / np.sum((x - mean_x) ** 2)
  b = mean_y - m * mean_x
  return m, b
```

```python
slope, intercept = simple_linear_regression(area, price)
```

```python
predictions = slope * area + intercept
```

```python
price - predictions
```

```python
plt.scatter(area, price, label = 'Actual Data')
plt.plot(area, predictions, color = 'red', label = 'Regression Line')
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()
```
